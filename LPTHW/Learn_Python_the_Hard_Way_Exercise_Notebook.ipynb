## Exercise 3 (Numbers and Math)

```
# Just printing normal string
print("I will now count my chickens:")
# Printing string and numeric value seperated with ',' 
# Moreover it automatically adds a space (' ') after the string
print("Hens", 25 + 30 / 6)
print("Roosters", 100 - 25 * 3 % 4)
# Just printing normal string
print("Now I will count the eggs:")
# I can see Python use PEMDAS
# Here first it calculates 4 % 2 = 0
# Then 1 / 4 = 0.25
# Then we have 3 + 2 + 1 - 5 + 0 - 0.25 + 6 = 6.75
print(3 + 2 + 1 - 5 + 4 % 2 - 1 / 4 + 6)
print("Is it true that 3 + 2 < 5 - 7?")
# Both the statement where first executed and there after checked
# 3 + 2 = 5
# 5 - 7 = -2
# Is 5 < -2 ?
print(3 + 2 < 5 - 7)
print("What is 3 + 2?", 3 + 2)
print("What is 5 - 7?", 5 - 7)
print("Oh, that's why it's False.")
print("How about some more.")
print("Is it greater?", 5 > -2)
print("Is it greater or equal?", 5 >= -2)
print("Is it less or equal?", 5 <= -2)
```

## Exercise 4 (Variables and Names)

```
cars = 100
space_in_a_car = 4.0
drivers = 30
passengers = 90
cars_not_driven = cars - drivers
cars_driven = drivers
carpool_capacity = cars_driven * space_in_a_car
average_passengers_per_car = passengers / cars_driven
print("There are", cars, "cars available.")
print("There are only", drivers, "drivers available.")
print("There will be", cars_not_driven, "empty cars today.")
print("We can transport", carpool_capacity, "people today.")
print("We have", passengers, "to carpool today.")
print("We need to put about", average_passengers_per_car,
"in each car.")
```

#### Write some variables that convert the inches and pounds to centimeters and kilograms

```
name = 'Debakar Roy'
height = 70.8 # inches
weight = 121.254 # lbs
height_cm = round(height*2.54, 2)
weight_kg = round(weight*0.453592, 2)
print(f"Let's talk about {name}.")
print(f"He's {height} inches tall i.e. {height_cm} centimeters tall.")
print(f"He's {weight} pounds heavy i.e. {weight_kg} kilograms heavy.")
```

## Exercise 6 (Strings and Text)

```
types_of_people = 10
# Using f-string to print '10' along with the string
x = f"There are {types_of_people} types of people."
binary = "binary"
do_not = "don't"
y = f"Those who know {binary} and those who {do_not}." # string inside string, 2 times
print(x)
print(y)
print(f"I said: {x}") # string inside string, 3rd one
print(f"I also said: '{y}'") # string inside string, 4th one
hilarious = False # This is boolean
joke_evaluation = "Isn't that joke so funny?! {}"
print(joke_evaluation.format(hilarious)) # using .format() syntax, used to apply a format to an already created string, such as in a loop
w = "This is the left side of..."
e = "a string with a right side."
print(w + e)
```

## Exercise 8 (Printing, Printing)

```
# This is one great example showing the use of .format() syntax
# We have used intgers, strings, booleans to show how it works
formatter = "{} {} {} {}"
print(formatter.format(1, 2, 3, 4))
print(formatter.format("one", "two", "three", "four"))
print(formatter.format(True, False, False, True))
# Note that formatter is itself a string
print(formatter.format(formatter, formatter, formatter, formatter))
print(formatter.format(
"Try your",
"Own text here",
"Maybe a poem",
"Or a song about fear"
))
```
